```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
# Dice 
Dice = [[u'ا',u'ن', u'ي', u'ه', u'ف' ,u'د'],
    [u'ا', u'ل', u'م', u'ب', u'ت' ,u'ذ'],
    [u'ا', u'ل', u'ن', u'ي', u'ت' ,u'ز'],
    [u'ا', u'ل', u'م', u'ه', u'ر' ,u'ع'],
    [u'ا', u'ن', u'و', u'ب', u'س' ,u'ص'],
    [u'ا', u'ل', u'ن', u'م', u'و' ,u'ي'],
    [u'ا', u'ل', u'ن', u'و', u'ه' ,u'ر'],
    [u'ا', u'ل', u'م', u'و', u'ث' ,u'غ'],
    [u'ا', u'ل', u'ن', u'ي', u'ف' ,u'ح'],
    [u'ا', u'ل', u'و', u'ه', u'ب' ,u'ة'],
    [u'ا', u'ن', u'ه', u'ع', u'ق' ,u'ض'],
    [u'ل', u'م', u'و', u'ي', u'ع' ,u'ق'],
    [u'ل', u'م', u'و', u'ي', u'ف' ,u'ق'],
    [u'ل', u'م', u'ب', u'ت', u'ط' ,u'ظ'],
    [u'ا', u'و', u'ي', u'ر', u'ت' ,u'ح'],
    [u'ا', u'ل', 'uن', u'م', u'ر' ,u'ش'],]

random.shuffle(Dice)

Choices = [random.choice(Dice[i]) for i in range(16)]

xpos = [0.125,0.375,0.625,0.875]
ypos = [0.125,0.375,0.625,0.875]



fig, ax = plt.subplots()

ax.axhline(y=0.25, xmin=0.0, xmax=1.0, color='r')
ax.axhline(y=0.5, xmin=0.0, xmax=1.0, color='r')
ax.axhline(y=0.75, xmin=0.0, xmax=1.0, color='r')

ax.axvline(x=0.25, ymin=0.0, ymax=1.0, color='r')
ax.axvline(x=0.5, ymin=0.0, ymax=1.0, color='r')
ax.axvline(x=0.75, ymin=0.0, ymax=1.0, color='r')

counter = 0
for x in range(4):
    for y in range(4):
        ax.text(xpos[x], ypos[y], Choices[counter], fontsize=30)
        counter = counter + 1

locs, labels = plt.xticks()           # Get locations and labels

plt.xticks(locs, [labels], alpha = 0)  # Set locations and labels

locs, labels = plt.yticks()           # Get locations and labels

plt.yticks(locs, [labels], alpha = 0)  # Set locations and labels


plt.show()
```
